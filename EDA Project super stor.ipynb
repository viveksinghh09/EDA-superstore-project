```python
break_even=set(df[df['Profit']==0]['Sub-Category'])
loss_making=set(df[df['Profit']<0]['Sub-Category'])
profit_making=set(df[df['Profit']>0]['Sub-Category'])
```

```python
for i in profit_making:
    if i in loss_making or i in break_even:
        pass
    else:
        print(i)
```

```python
profit_subcat=[]
for i in profit_making:
    if i in loss_making or i in break_even:
        pass
    else:
        profit_subcat.append(i)
        
```

```python
print(profit_subcat)
```

```python
round(float(df[df['Sub-Category'].isin(profit_subcat)]['Profit'].mean()),2)
```
